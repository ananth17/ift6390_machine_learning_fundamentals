```python
# preprocessing functions

# - deskewing
# - ink normalization

from scipy.spatial import distance
from scipy.ndimage import interpolation


def ink_normalize(x):
    return x /np.linalg.norm(x)

def moments ( image ) :
    # taken from 
    c0,c1 = mgrid[:image.shape[0],:image.shape[1]]
    imgsum = np.sum(image)
    m0 = np.sum(c0*image )/imgsum
    m1 = np.sum(c1*image )/imgsum
    m00 = np.sum(( c0 - m0 )*( c0 - m0 )*image)/ imgsum
    m11 = np.sum(( c1 - m1 )*( c1 - m1 )*image)/ imgsum
    m01 = np.sum(( c0 - m0 )*( c1 - m1 )*image)/ imgsum
    return array([ m0, m1]), np.array([[m00, m01], [m01, m11]])


def deskew(image):
    c,v = moments(image)
    alpha = v[0,1]/v[0,0]
    affine = array([[1,0],[alpha,1]])
    ocenter = array(image.shape)/2.0
    offset = c-dot(affine,ocenter)
    return interpolation.affine_transform(image,affine,offset=offset)


def preprocess(digit):
    # deskew and then ink normalize
    #assert digit.shape == (784,)
    return np.ndarray.flatten(ink_normalize(deskew(np.reshape(digit, (28, 28)))))
    
```

```python
# check the overall accuracy
def accuracy(data, target, classifier):
    return np.sum(np.equal(classifier.predict(data), target)) / float(len(target))

# check the class-specific stats
def get_stats(class_to_data, classifier):
    # per-class misclassification rate
    # per-class misclassification identities
    class_accuracy = dict()
    class_destination = dict()
    
    for (c, data) in class_to_data.iteritems():
        predicted = classifier.predict(data)
        predictions_length = float(len(predicted))
        
        class_accuracy[c] = np.sum(map(lambda x:x==c, predicted)) / predictions_length
        
        class_destination[c] = dict()
        
        # initialize the destinations with 0
        for d in class_to_data.keys():
            class_destination[c][d] = 0
        
        for predicted_class in predicted:
            class_destination[c][predicted_class] += 1

    return (class_accuracy, class_destination)

print accuracy(train_x, train_y, lin_classifier)
print accuracy(test_x, test_y, lin_classifier)

train_accuracy, train_classification = get_stats(train_data, lin_classifier)
test_accuracy, test_classification = get_stats(test_data, lin_classifier)
```

```python
def heatmap_classification(classification, name):
    data = np.empty((0, 10))
    for i in range(0, 10):
        row = np.zeros(10)
        for j in range(0, 10):
            row[j] = classification[i][j]
        row /= float(np.sum(row))
        data = np.vstack((data, row))
    figure(figsize = (10, 10))
    plt.title(name)
    plt.ylabel("image class")
    plt.xlabel("predicted class")
    plt.xticks(range(0,10))
    plt.yticks(range(0,10))
    plt.imshow(data, cmap=plt.cm.summer_r, interpolation='none')
    plt.colorbar()
    plt.show()
    class_accuracy = []
    for i in range(0, 10):
        class_accuracy.append(data[i][i])
    figure(figsize = (9, 7))
    plt.bar(range(0,10), class_accuracy)
    plt.title('accuracy per class')
    plt.xticks(range(0,10))
    plt.show()
    return data

train_heatmap = heatmap_classification(train_classification, "classifications in training set")
test_heatmap  = heatmap_classification(test_classification, "classification in test set")
```

```python
def distort_random_grad(digit, original_class, target, classifier, seed=42, step=25, epsilon=0.1):
    #
    np.random.seed(seed)
    quantum = 1./256.
    
    
    distortion = np.zeros(len(digit))
    gradients = lin_classifier.coef_[target] - lin_classifier.coef_[original_class]
    abs_gradients = np.abs(gradients)
    hard_threshold = 100000  # will stop in case of failure
    number_modifications = 0
    while number_modifications < hard_threshold:
        index = np.random.randint(low=0, high=len(digit))
        gradient = np.random.choice([-1., 1.]) if gradients[index] == 0 else np.sign(gradients[index])
        if abs_gradients[index] < epsilon:
            continue
        distortion[index] += quantum * gradient
        number_modifications += 1
        if number_modifications % step == 0:
            predicted_class = classifier.predict(digit+distortion)
            if predicted_class != original_class:
                #print predicted_class
                #print number_modifications
                break
    
    return (distortion, np.linalg.norm(distortion)**2)

c1=0
digit_index = 0
n_per_class = 10
distortions = []
distances = []
targets = []
digit = train_data[c1][digit_index]
show_mnist_digit(digit)
for c2 in range(0, 10):
    if c2 == c1:
        continue

    for _ in range(0, n_per_class):
        distortion, distance = distort_random_grad(digit, c1, c2, lin_classifier, seed=np.random.randint(low=0, high=100000))
        distortions.append(distortion)
        distances.append(distance)
        targets.append(c2)
        #plt.imshow(np.reshape(np.abs(distortion), (28, 28)), cmap=plt.cm.gray_r, interpolation='none')
        #plt.colorbar()
        #plt.show()
```
